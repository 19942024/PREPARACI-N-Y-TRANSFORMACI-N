### Ejercicio 1
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Escribe una función (llámala "nulls_number") que reciba como argumento un DataFrame Pandas, obtenga las 5 columnas con mayor número de valores nulos y devuelva una serie Pandas conteniendo los nombres de dichas columnas como índices y el número de nulos como valores.
</div>

```python
# Definimos la función nulls_number
def nulls_number(df):
    # Calculamos el número de nulos con la función isnull
    nulos = df.isnull().sum()

    # Ordenamos las columnas de forma descendente y selecciona las 5 primeras columnas
    nulos = nulos.sort_values(ascending=False).head(5)

    return nulos
```

### Ejercicio 2
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    <p>El objetivo de este ejercicio es crear una función (a la que vamos a llamar "mask") que reciba dos argumentos: una imagen (con formato NumPy) y un umbral (valor real entre 0 y 1), y que devuelva otro array NumPy (otra imagen) que muestre puntos de color negro allí donde la imagen original "no sea suficientemente clara". Es decir, queremos obtener una máscara de la imagen.</p>
    <p>Para ello se recomienda que la función:</p>
    <ul>
        <li>Compruebe si la imagen tiene tres o cuatro capas. En el caso de tener cuatro, la última (la capa de transparencia) deberá eliminarse.</li>
        <li>Calcule el valor medio de las tres capas RGB (es decir, que cree un array con el ancho y alto de la imagen original en el que cada punto sea el valor medio de los tres valores RGB para dicho pixel</li>
        <li>Por último, devuelva un array con las mismas dimensiones que el array generado en el paso anterior que contenga el valor 1 cuando el valor correspondiente en el array anterior sea un número superior al umbral indicado (segundo argumento pasado a la función), y 0 en caso contrario.</li>
    </ul>
</div>

```python
# Definimos la función 

def mask(imagen, umbral):
   
    # Eliminamos el canal alfa si está presente.
    if imagen.shape[2] == 4:
        imagen = imagen[:, :, :3]

    # Calculamos el valor medio de las 3 capas RGB para cada píxel.
    mean_rgb = np.mean(imagen, axis=2)

    # Normalizar los valores medios para estar en el rango de 0 a 1 si no lo están.
    if imagen.max() > 1:
        mean_rgb = mean_rgb / 255   # dividimos para el valor maxiomo de pixel 

    # Crear la máscara aplicando el umbral.
    #comprueba si cada valor en mean_rgb es mayor que el umbral. Si es verdadero se asigna un 1, si es falso, asigna un 0.
    mask = np.where(mean_rgb > umbral, 1, 0)

    return mask
```
